```python
import numpy as np

def find_zero_sum_rows_cols(matrix):
   

    row_sums = np.sum(matrix, axis=1)  

    column_sums = np.sum(matrix, axis=0)

    zero_sum_rows = np.where(np.isclose(row_sums, 0))[0]
    
    zero_sum_columns = np.where(np.isclose(column_sums, 0))[0]
    
    return zero_sum_rows, zero_sum_columns


matrix = new_matrixH
zero_sum_rows, zero_sum_columns = find_zero_sum_rows_cols(matrix)


print(f"Number of rows with sum 0: {len(zero_sum_rows)}")
print(f"Rows with sum 0: {zero_sum_rows.tolist()}")
print(f"Number of columns with sum 0: {len(zero_sum_columns)}")
print(f"Columns with sum 0: {zero_sum_columns.tolist()}")
```

```python
import numpy as np

def find_zero_sum_columns(matrix):
    matrix = np.array(matrix)  # Ensure the input is converted to a NumPy array

    # Calculate column sums
    column_sums = np.sum(matrix, axis=0)
    
    # Find columns with sum 0
    zero_sum_columns = np.where(column_sums == 0)[0]
    
    return zero_sum_columns

# Example matrix
matrix = new_matrixH

# Find zero-sum columns
zero_sum_columns = find_zero_sum_columns(matrix)

# Print the results
print(f"Number of columns with sum 0: {len(zero_sum_columns)}")
print(f"Columns with sum 0: {zero_sum_columns.tolist()}")
```
